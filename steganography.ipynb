## Hiding in plain sight: image steganography

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

## Compare two images side by side

```python
tower_image = plt.imread('original_images/tower.tiff')
hidden_image = plt.imread('altered_images/tower.tiff')

fig, ax = plt.subplots(1, 2, figsize=(30, 10))
ax[0].imshow(tower_image)
ax[0].set_title('A normal image')
ax[1].imshow(hidden_image)
ax[1].set_title('A special image')

[axi.set_axis_off() for axi in ax.ravel()];
```

## Do you see anything?

```python
from math import ceil

bits = 4
step = ceil(255.0 / bits)

tower_remainder = (tower_image % bits * step).astype(np.uint8)
hidden_remainder = (hidden_image % bits * step).astype(np.uint8)

fig, ax = plt.subplots(1, 2, figsize=(20, 10))
ax[0].imshow(tower_remainder)
ax[1].imshow(hidden_remainder)

[axi.set_axis_off() for axi in ax.ravel()];
```
