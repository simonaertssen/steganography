## Hiding in plain sight: image steganography

```python
%matplotlib inline
from numpy import uint8
import matplotlib.pyplot as plt
```

# Abundancy of information in images
Images contain an abundancy of information.

Example: JLI's Hot End Tableware Inspection (HETI) =
Inspection of glassware coming out of a glass extrusion oven.

Detect:
 - defects
 - asymmetries


```python
heti = plt.imread('original_images/heti.jpg')
plt.figure(figsize=(20, 10))
plt.imshow(heti)
plt.axis('off');
```

## Compare two images side by side

```python
tower_image = plt.imread('original_images/tower.tiff')
hidden_image = plt.imread('altered_images/tower.tiff')

fig, ax = plt.subplots(1, 2, figsize=(30, 15))
ax[0].imshow(tower_image)
ax[0].set_title('A normal image', fontsize=16)
ax[1].imshow(hidden_image)
ax[1].set_title('A special image', fontsize=16)

[axi.set_axis_off() for axi in ax.ravel()];
```

## Do you see anything?

```python
from math import ceil

bits = 18
step = 255.0 / bits

tower_remainder = (tower_image % bits * step).astype(uint8)
hidden_remainder = (hidden_image % bits * step).astype(uint8)

fig, ax = plt.subplots(1, 2, figsize=(30, 15))
ax[0].imshow(tower_remainder)
ax[0].set_title('A normal image', fontsize=16)
ax[1].imshow(hidden_remainder)
ax[1].set_title('A special image', fontsize=16)

[axi.set_axis_off() for axi in ax.ravel()];
```
